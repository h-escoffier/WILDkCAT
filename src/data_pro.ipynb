```python
import pickle
import pandas as pd

def load_pckl_data(path):
    with open(path, 'rb') as f:
        obj = pickle.load(f)
    return obj

kcat_data = load_pckl_data('../data/kcat/Uniprot_kcat.pkl')
print(kcat_data.shape)
# print(kcat_data.head())

kcat_data.to_csv('../data/kcat/Uniprot_kcat.tsv', sep='\t', index=False)
```
